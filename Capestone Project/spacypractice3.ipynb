Entity Ruler - used to on how to think of a set of rules based on a list of known things or a set of rules generated by RegEx code or Linguistic features

Machine Learning- is used when you don't know what are the set of rules are for e.g. names of people - u need to generate a robust feature

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
text = "West Chestertenfieldville was referenced  in Mr. Deeds."
```

```python
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
ruler = nlp.add_pipe("entity_ruler") #to define the entity
```

```python
#add entity ruler to the nlp model
nlp.analyze_pipes()
```

```python
#adding patterns, spacy model is going to look for and the labels is going to assign 
patterns = [
    {"label": "GPE", "pattern": "West Chestertenfieldville"}
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc2 = nlp(text)
for ent in doc2.ents:
    print(ent.text, ent.label_)
```

```python
nlp2 = spacy.load("en_core_web_md")
```

```python
ruler = nlp2.add_pipe("entity_ruler", before = "ner")
```

```python
ruler.add_patterns(patterns)
```

```python
doc = nlp2(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
nlp2.analyze_pipes()
# entity ruler dominates the ner
```

```python
nlp3 = spacy.load("en_core_web_md")
```

```python
ruler = nlp3.add_pipe("entity_ruler", before = "ner")
```

```python
patterns = [
    {"label": "GPE", "pattern": "West Chestertenfieldville"},
    {"label": "FILM", "pattern": "Mr. Deeds"}
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc = nlp3(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
